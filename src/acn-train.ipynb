```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from pycaret.regression import *

import shutup; shutup.please()
```

```python
data = pd.read_csv('../input/acn-proj/tcp_data.csv')
```

```python
test = data.drop(data['new_window_size'])
```

```python
reg = setup(data, target = 'new_window_size', session_id=123, log_experiment=True, experiment_name='ModelsforNet',use_gpu=True)
```

```python
best_model = compare_models(fold=6,turbo=False)
```

```python
tuned_cat = tune_model(best_model, n_iter=100, optimize = 'MAE')
```

```python
pred_holdouts = predict_model(tuned_cat)
```

```python
plot_model(tuned_cat, plot = 'parameter')
```

```python
save_model(tuned_cat, model_name='best-model')
```

```python
load_saved_model = load_model('best-model')
```
